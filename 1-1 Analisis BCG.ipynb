```r
library(xlsx)
library(ggplot2)
library(ggrepel)
```

```r
df_bcg <- read.xlsx('Soal Latihan BCG.xlsx', sheetName = 1)
```

```r
df_bcg
```

```r
median(df_bcg$salesGrowth)
```

```r
median(df_bcg$marketPosition)
```

```r
df_bcg$margin = df_bcg$profit / df_bcg$directCost
```

```r
plot <- ggplot(df_bcg) + geom_point(aes(x = marketPosition, y = log(salesGrowth), size = yearlySales, color = margin)) + geom_segment(aes(x = 0.5, y = log(0.08), xend = 4.5, yend = log(0.08))) + geom_segment(aes(x = 2.5, y = log(0.015), xend = 2.5, yend = log(0.5))) + scale_size(range = c(2, 13)) + scale_colour_gradient2(low = 'darkred', mid = 'red', high = 'green')
```

```r
plot
```

```r
labelPlot <- plot + geom_label(aes(x = marketPosition, y = log(salesGrowth), label = produk), nudge_x = 0.29)
```
